```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import ensemble
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.externals import joblib
```

```python
data = pd.read_excel('Reference_Value.xlsx')
data = data.dropna()
real_data = pd.read_csv('test.csv')
real_data = real_data.dropna()
```

```python
train_data, test_data = train_test_split(data, test_size=0.20)
train_data = np.array(train_data)
test_data = np.array(test_data)
real_data = np.array(real_data)
```

```python
n_est = 10000
params = {
    'n_estimators': n_est,
    'max_depth': 10,
    'min_samples_split': 2,
    'learning_rate': 0.001, 'loss': 'ls'}
clf = ensemble.GradientBoostingRegressor(**params)
```

```python
col = 24
y_train = []
X_train = []
for i in range(len(train_data)):
    y_train.append(train_data[i][col])
    X_train.append(np.append(train_data[i][:col],train_data[i][(col+1):]))
y_test = []
X_test = []
for i in range(len(test_data)):
    y_test.append(test_data[i][col])
    X_test.append(np.append(test_data[i][:col],test_data[i][(col+1):]))
y_real = []
X_real = []
for i in range(len(real_data)):
    y_real.append(real_data[i][col])
    X_real.append(np.append(real_data[i][:col],real_data[i][(col+1):]))
```

```python
clf.fit(X_train,y_train)
filename = 'n_%d.sav' %n_est
joblib.dump(clf, filename)
```

```python
predict_train = clf.predict(X_train)
predict_test = clf.predict(X_test)
plt.figure(figsize=(100,20))
plt.subplot(211)
plt.title('Training MSE: {}'.format(mean_squared_error(y_train, predict_train)))
plt.plot(range(len(y_train)), predict_train, 'r--', range(len(y_train)), y_train, 'b--')
plt.savefig('train.png')
print("Training MSE: {}".format(mean_squared_error(y_train, predict_train)))

plt.subplot(212)
plt.title('Testing MSE: {}'.format(mean_squared_error(y_test, predict_test)))
plt.plot(range(len(y_test)),predict_test,'r--',range(len(y_test)),y_test,'b--')
plt.savefig('result/test_%d.png' % n_est)
print("Testing MSE: {}".format(mean_squared_error(y_test, predict_test)))
print("Training max: %.3f / Traing min: %.3f" % (max(predict_train),min(predict_train)))

rec = []
rec.append("n_estimator %d" % n_est)
rec.append(" Training MSE: %.8f" % mean_squared_error(y_train, predict_train))
rec.append(" Testing MSE: %.8f" % mean_squared_error(y_test, predict_test))
rec.append("\n")
fh = open("result/test.txt", "a")
fh.writelines(rec) 
fh.close()
```

```python
predict = clf.predict(X_real)
plt.figure(figsize=(100,20))
plt.subplot(211)
plt.title('Real value')
plt.plot(range(len(y_real)), predict, 'r--', range(len(y_real)), y_real, 'b--')

threshold = 30
residual = predict-y_real
anomaly = [np.nan]*len(y_real)
if(threshold > 0):
    anomaly = ((residual > threshold) | (residual < (-1*threshold))).astype(int)
    anomaly = anomaly * max(residual)
    anomaly[anomaly==0] = np.nan

plt.subplot(212)
plt.title('Residual')
plt.plot(range(len(y_real)), residual,'r--' , range(len(y_real)),anomaly,'rx')
plt.savefig('result/real_%d.png' % n_est)
print("Input max: %3.f / Input min: %.3f" % ((max(y_real),min(y_real))))
print("Predict max: %.3f / Predict min: %.3f" % (max(predict),min(predict)))
```
